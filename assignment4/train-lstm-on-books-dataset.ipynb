# Import necessary libraries and packages

```python
import os
import re
```

# Inspect each file of each book

```python
path_to_data_files = "/kaggle/input/10000-vietnamese-books/output"

# Get all the file names inside a list
filenames = os.listdir(path_to_data_files)
filenames[:10]
```

```python
filepaths = [path_to_data_files + '/' + filename for filename in filenames]
filepaths[:10]
```

```python
def preprocess_text(text: str) -> str:
    # Remove all &quote;
    text = text.replace("&quote;", "")
    
    # Replace multiple spaces, tabs, newlines with a single space
    text = re.sub(r'\s+', ' ', text)
    
    # Remove leading and trailing spaces
    text = text.strip()
    
    return text
```

```python
texts = []
for filepath in filepaths:
    with open(filepath,
             mode="r",
             encoding="utf-8") as file:
        raw_text = file.read()
        preprocessed_text = preprocess_text(raw_text)
        texts.append(preprocessed_text)
```
